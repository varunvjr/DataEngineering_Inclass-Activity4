```python
# Intra-Record Assertion
# Example-2 Summation Of Latitude Degree, Latitude Minute and Latitude Second for Each Record
#The sum of each record latitude degree,latitude minute and latitude second must be in the range of (41+0+0.00) and (47+59+59.99)

import pandas as pd
import numpy as np
import math

crash_data=pd.read_csv("./crash.csv")
crash_data=crash_data.dropna(subset=["Latitude Degrees","Latitude Minutes","Latitude Seconds"])
degrees=crash_data["Latitude Degrees"]
minutes=crash_data["Latitude Minutes"]
seconds=crash_data["Latitude Seconds"]
data={"Degrees":degrees,"Minutes":minutes,"Seconds":seconds}
df= pd.DataFrame(data)
lowLimit=41.00+0+0.0
highLimit=47+59+59.99
for row in df.iterrows():
    validData=row[1]["Degrees"]+row[1]["Minutes"]+row[1]["Seconds"]
    if validData>=lowLimit and validData<=highLimit:
        print("No Violation")
    else:
        print("Data Violation")

```

```python
# Inter Record Assertion

# Example 1: Every Crash  Recorded with record type=1 have the record type=2 and record type=3

import pandas as pd
import numpy as np

df=pd.read_csv("./crash.csv")
CrashesDF = df[df['Record Type'] == 1]
VehiclesDF = df[df['Record Type'] == 2]
ParticipantsDF = df[df['Record Type'] == 3]

CrashesDF = CrashesDF.dropna(axis=1,how='all')
VehiclesDF = VehiclesDF.dropna(axis=1,how='all')
ParticipantsDF = ParticipantsDF.dropna(axis=1,how='all')
crashesDF=CrashesDF.drop_duplicates(["Crash ID"])
vehiclesDF=VehiclesDF.drop_duplicates(["Crash ID"])
ParticipantsDF=ParticipantsDF.drop_duplicates(["Crash ID"])
print(crashesDF.shape[0])
print(vehiclesDF.shape[0])
print(ParticipantsDF.shape[0])

```

```python
# Inter Record Assertion

# Example 2: Every crash ID must have atleast 1 vehicle Id associated with it
VehicleCrash = VehiclesDF[['Crash ID', 'Vehicle ID']]
VehicleCrash = VehicleCrash.groupby(['Crash ID'])
VehicleCrash = VehicleCrash.count()
for i in VehicleCrash['Vehicle ID']:
    if(i <= 0):
        print("Failed")
print("Passed")
```

```python
# Summary Assertion
# Example 1: Summation of Latitude Degree  
# The sum of all the recordâ€™s latitude degrees must range between 41*numRows and 47*numRows. (numRows is the number of unique records with the given crash id)

import pandas as pd
import numpy as np
import math

crash_data=pd.read_csv("./crash.csv")
crash_data=crash_data.dropna(subset=["Latitude Degrees"])
numRows=crash_data.shape[0]
if sum(crash_data["Latitude Degrees"])>=numRows*41 and sum(crash_data["Latitude Degrees"])<=numRows*47:
    print("Pass")
else:
    print("Data Violation")
```

```python
# # Statistical Distribution Assertion
# Example 2: Number of Crashes occured in all the months of the year

import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline



df=pd.read_csv("./crash.csv")
CrashesDF = df[df['Record Type'] == 1]
VehiclesDF = df[df['Record Type'] == 2]
ParticipantsDF = df[df['Record Type'] == 3]

CrashesDF = CrashesDF.dropna(axis=1,how='all')
VehiclesDF = VehiclesDF.dropna(axis=1,how='all')
ParticipantsDF = ParticipantsDF.dropna(axis=1,how='all')
MonthlyCrashes=CrashesDF[["Crash ID","Crash Month"]]
MonthlyCrashes=MonthlyCrashes.groupby(["Crash Month"])
crashCount=MonthlyCrashes.count()
Yaxis=crashCount["Crash ID"].tolist()
Xaxis=["Jan","Feb","March","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
df=pd.DataFrame({'Months':Xaxis,"MonthlyCrashes":Yaxis})
ax=df.plot.bar(x='Months',y="MonthlyCrashes",rot=0)



```
